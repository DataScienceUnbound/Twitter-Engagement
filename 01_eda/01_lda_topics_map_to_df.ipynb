```python
import pandas as pd
import numpy as np
```

```python
image_df = pd.read_csv("image_data_w_lda_topics.csv")
```

```python
image_df.head()
```

```python
image_df = image_df.drop("Unnamed: 0", axis =1)
```

```python
a = max(image_df[(image_df["img"]=="img_00010")&(image_df["img_class"] == "media")]["label_topic_prob"])
```

```python
df = pd.DataFrame(image_df.loc[(image_df["img"]=="img_00010") & 
                                     (image_df["img_class"] == "media") & 
                                     (image_df["label_topic_prob"]==a)])
```

```python
for img, img_class in zip(image_df["img"], image_df["img_class"]):
    x = max(image_df[(image_df["img"]==img)&(image_df["img_class"] == img_class)]["label_topic_prob"])
    a_df = pd.DataFrame(image_df.loc[(image_df["img"]==img) & 
                                     (image_df["img_class"] == img_class) & 
                                     (image_df["label_topic_prob"]==x)])
    df = df.append(a_df)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df=df.merge(pd.get_dummies(df["label_topic"]).drop(14.0, axis =1), left_index=True, right_index=True)
```

```python
df.drop("label_topic", axis=1)
```

```python
import math

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import CountVectorizer

from nltk.stem import WordNetLemmatizer
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
```

```python
img_merge['brand_name'] = img_merge.tweet_url.apply(lambda x: x.replace('https://www.twitter.com/', '')[:-26])
```

```python
img_merge.timestamp = pd.to_datetime(img_merge.timestamp)
```

```python
img_merge['hour'] = img_merge.timestamp.apply(lambda x: x.hour)
img_merge['weekday'] = img_merge.timestamp.apply(lambda x: x.weekday_name)
img_merge['month'] = img_merge.timestamp.apply(lambda x: x.month)
```

```python
img_merge.hour = img_merge.hour.astype('category')
img_merge.month = img_merge.month.astype('category')
```

```python
img_merge['uploaded_media'] = img_merge["media_url"].notnull()
img_merge.uploaded_media = img_merge.uploaded_media.apply(lambda x: 1 if x==True else 0)
img_merge['text_only'] = img_merge.type.apply(lambda x: 1 if x =="['text only']" else 0)
```

```python
img_merge['question'] = img_merge.text.apply(lambda x: 1 if '?' in x else 0)
```

```python
img_merge['has_hashtag'] = img_merge.hashtags.notnull()
img_merge.has_hashtag = img_merge.has_hashtag.apply(lambda x: 1 if x==True else 0)
```

```python
img_merge.to_csv('all_data_feats_eng.csv')
```
